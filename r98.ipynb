```python
%matplotlib inline
from astropy.cosmology import LambdaCDM, FlatLambdaCDM
import astropy.units as u
from scipy.optimize import minimize
import numpy as np
import pylab as plt
```

```python
def lnlike(theta, z, mu, muerr):
    # the log(likelihood)
    # theta is the free parameters, z is the redshift
    # mu/muerr are distance moduli/uncertainties
    Om,b = theta
    # Om is the cosmic matter density, b is just an offset
    # b is necessary because we don't know the SN distance
    # matter cannot be negative
    model = LambdaCDM(H0=70,Om0=Om,Ode0=0,Tcmb0=2.725 * u.K)
    # H0 is the current expansion rate of the universe
    # Ode is dark energy - none for now
    inv_sigma2 = 1.0/muerr**2.
    return -0.5*(np.sum((mu-model.distmod(z).value+b)**2*inv_sigma2 - np.log(inv_sigma2)))
```

```python
z,mu,muerr = np.loadtxt('riess98.dat',unpack=True,usecols=[1,2,3]) # a text file with supernova data
plt.errorbar(z,mu,yerr=muerr,fmt='o')
plt.xlabel('redshift')
plt.ylabel('log(distance)')
```

```python
nll = lambda *args: -lnlike(*args)
result = minimize(nll, [0.6, 0.0 ], args=(z, mu, muerr),bounds=((0,1),(None,None)),method='SLSQP')
m_ml, b_ml = result["x"]
```

```python
print(result['message'])
print('Omega Matter = %.3f'%m_ml)
```

```python
bestmodel = FlatLambdaCDM(H0=70,Om0=m_ml,Tcmb0=2.725 * u.K)
zmod = np.arange(0,1,0.01)
plt.plot(zmod,bestmodel.distmod(zmod).value+b_ml,label='cosmological model')
plt.errorbar(z,mu,yerr=muerr,fmt='o',label='SN Ia data')
plt.xlabel('redshift')
plt.ylabel('log(distance)')
plt.legend()
```

```python
plt.errorbar(z,mu-bestmodel.distmod(z).value+b_ml,yerr=muerr,fmt='o',label='SN Ia data')
plt.xlabel('redshift')
plt.ylabel('distance-model')
plt.axhline(0,color='k')
plt.title('Residual Diagram, $\Omega_{M}$=0')
```

```python
# The best fit says there's no matter in the universe!
# That means the universe is moving faster than we can explain using matter alone
# Let's try adding dark energy, and forcing matter to be closer to what we measure
# from galaxy cluster counting
def lnlikeDE(theta, z, mu, muerr):
    # the log(likelihood)
    # theta is the free parameters, z is the redshift
    # mu/muerr are distance moduli/uncertainties
    Om,b = theta
    if Om <= 0 or Om > 1: return -np.inf
    # Om is the cosmic matter density, b is just an offset
    # b is necessary because we don't know the SN distance
    # Ode is the dark energy density
    # matter cannot be negative
    model = FlatLambdaCDM(H0=70,Om0=Om,Tcmb0=2.725 * u.K)
    # H0 is the current expansion rate of the universe
    # Ode is dark energy - none for now
    inv_sigma2 = 1.0/muerr**2.
    return -0.5*(np.sum((mu-model.distmod(z).value+b)**2*inv_sigma2 - np.log(inv_sigma2)))
```

```python
nllDE = lambda *args: -lnlikeDE(*args)
result = minimize(nllDE, [0.6, 0.0 ], args=(z, mu, muerr),bounds=((0,1),(None,None)),method='SLSQP')
m_ml, b_ml = result["x"]
```

```python
bestmodelDE = FlatLambdaCDM(H0=70,Om0=m_ml,Tcmb0=2.725 * u.K)
plt.errorbar(z,mu-bestmodelDE.distmod(z).value+b_ml,yerr=muerr,fmt='o',label='SN Ia data')
plt.xlabel('redshift')
plt.ylabel('distance-model')
plt.axhline(0,color='k')
plt.title('Residual Diagram, $\Omega_{M}$=%.2f, $\Omega_{\Lambda}$=%.2f'%(m_ml,1-m_ml))
```

```python
ndim, nwalkers = 2, 100
pos = [result["x"] + 1e-4*np.random.randn(ndim) for i in range(nwalkers)]
import emcee
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnlikeDE, args=(z, mu, muerr))
```

```python
sampler.run_mcmc(pos, 500)
```

```python
samples = sampler.chain[:, 50:, :].reshape((-1, ndim))
for m, b in samples[np.random.randint(len(samples), size=100)]:
    testmodel = FlatLambdaCDM(H0=70,Om0=m,Tcmb0=2.725 * u.K)
    plt.plot(z, testmodel.distmod(z).value+b-bestmodelDE.distmod(z).value-b_ml, 
            color="k", alpha=0.1)
plt.errorbar(z,mu-bestmodelDE.distmod(z).value+b_ml,yerr=muerr,fmt='o',label='SN Ia data')
plt.xlabel('redshift')
plt.ylabel('distance-model')
plt.axhline(0,color='r')
plt.title('Residual Diagram, $\Omega_{M}$=%.2f, $\Omega_{\Lambda}$=%.2f'%(m_ml,1-m_ml))
```

```python
m_mcmc, b_mcmc = map(lambda v: (v[1], v[2]-v[1], v[1]-v[0]),
                                 zip(*np.percentile(samples, [16, 50, 84],
                                                    axis=0)))
print('Omega Lambda = %.3f +/- %.3f'%(1-m_mcmc[0],np.mean([m_mcmc[1],m_mcmc[2]])))
```
